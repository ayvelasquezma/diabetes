```python
X_train, X_test, y_train, y_test = train_test_split(X_diabetes,
                                                    y_diabetes,
                                                    test_size=0.3,
                                                    stratify=y_diabetes,
                                                    random_state=random)
print(f'Número de muestras en entrenamiento: {X_train.shape[0]}')
print(f'Número de muestras en prueba: {X_test.shape[0]}')
print(f'Número de características: {X_train.shape[1]}')
```

```python
param_grid_knn = {'n_neighbors':     [i for i in range(1, 42, 2)],
                  'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']}

grid_knn = GridSearchCV(KNeighborsClassifier(),
                   param_grid=param_grid_knn,
                   verbose=1,
                   return_train_score=True
                   )
grid_knn.fit(X_train, y_train)
```

```python
print(grid_knn.best_params_)
print(grid_knn.best_score_)
```
